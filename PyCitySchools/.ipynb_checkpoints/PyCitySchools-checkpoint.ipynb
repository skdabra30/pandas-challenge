```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total Number of Schools
school_count = len(school_data_complete_df["school_name"].unique())
print(school_count)

# Total Number of Schools
student_count = school_data_complete_df["Student ID"].count()
print(student_count)

#Total budget
total_budget = school_data["budget"].sum()
print(total_budget)

#Average Math Score
average_math = school_data_complete_df["math_score"].mean()
print(average_math)

#Average Reading Score
average_reading = school_data_complete_df["reading_score"].mean()
print(average_reading)

#Percentage of students with a math score of 70 or greater
math_passing_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]
math_passing_percentage = math_passing_count / student_count *100
print(math_passing_percentage)

#Percentage of students with a reading score of 70 or greater
reading_passing_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
reading_passing_percentage = reading_passing_count / student_count *100
print(reading_passing_percentage)

#Percentage of students with a math and reading score of 70 or greater
both_passing_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
both_passing_percentage = both_passing_count / student_count *100
print(both_passing_percentage)
```

```python
# Create dataframe for District Summary
district_summary_dicts=[{"Total Schools":school_count,"Total Students":student_count,"Total Budget":total_budget,
                         "Average Math Score":average_math,"Average Reading Score":average_reading,
                         "Passed Math (%)":math_passing_percentage,"Passed Reading (%)":reading_passing_percentage,
                         "Passed Both (%)":both_passing_percentage}]
district_summary_df = pd.DataFrame(district_summary_dicts)
district_summary_df
```

```python
# Clean up dataframe
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.2f}".format)
district_summary_df["Passed Math (%)"] = district_summary_df["Passed Math (%)"].map("{:,.1f}%".format)
district_summary_df["Passed Reading (%)"] = district_summary_df["Passed Reading (%)"].map("{:,.1f}%".format)
district_summary_df["Passed Both (%)"] = district_summary_df["Passed Both (%)"].map("{:,.1f}%".format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Determine school type and index school name
school_type = school_data.set_index(["school_name"])["type"]
school_type.head()
```

```python
# Determine students per school
school_students = school_data.set_index(["school_name"])["size"]
school_students.head()
```

```python
# Determine Budget per school
school_budget = school_data.set_index(["school_name"])["budget"]
school_budget.head()
```

```python
# Determine budget per student
school_budget_per_student=school_budget/school_students
school_budget_per_student.head()
```

```python
# Determine average math score
school_math=school_data_complete_df.groupby(["school_name"]).mean()["math_score"]
school_math.head()
```

```python
# Determine average reading score
school_reading=school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]
school_reading.head()
```

```python
# Determine % passing math
school_passing_math = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)]
school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"]
school_passing_math = school_passing_math / school_students * 100
school_passing_math.head()
```

```python
# Determine % passing reading
school_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)]
school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"]
school_passing_reading = school_passing_reading / school_students * 100
school_passing_reading.head()
```

```python
# Determine % passing both
school_passing_both = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)
    & (school_data_complete_df["math_score"] >= 70)]
school_passing_both = school_passing_both.groupby(["school_name"]).count()["student_name"]
school_passing_both = school_passing_both/school_students * 100
school_passing_both.head()
```

```python
# Create dataframe for School Summary
school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": school_students,
    "Total School Budget": school_budget,
    "Per Student Budget": school_budget_per_student,
    "Average Math Score": school_math,
    "Average Reading Score": school_reading,
    "% Passing Math": school_passing_math,
    "% Passing Reading": school_passing_reading,
    "% Overall Passing": school_passing_both})

school_summary_df.head()
```

```python
# Make a copy of the dataframe for future use
school_summary_spending_df=school_summary_df.copy()
school_summary_size_df=school_summary_df.copy()
school_summary_type_df=school_summary_df.copy()
school_summary_spending_df.head()
school_summary_size_df.head()
school_summary_type_df.head()
```

```python
# Clean up dataframe
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:,.2f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("${:,.2f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("${:,.2f}".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("${:,.2f}".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("${:,.2f}".format)
school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create Bins and group names
spending_bins=[0, 580, 610, 640, 670]
group_names=["$0-$579","$580-$609","$610-$639","$640-$670"]
```

```python
# Cut in per student budget
school_summary_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_budget_per_student, spending_bins, labels=group_names)
school_summary_spending_df.head()
```

```python
school_summary_spending_df.dtypes
```

```python
# Calculate averages
math_scores_spending = school_summary_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

reading_scores_spending = school_summary_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

passing_math_spending = school_summary_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

passing_reading_spending = school_summary_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

passing_both_spending = school_summary_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Create Dataframe
spending_summary_df = pd.DataFrame({
          "Average Math Score" : math_scores_spending,
          "Average Reading Score": reading_scores_spending,
          "% Passing Math": passing_math_spending,
          "% Passing Reading": passing_reading_spending,
          "% Overall Passing": passing_both_spending})
spending_summary_df
```

```python
spending_summary_df.dtypes
```

```python
# Clean Dataframe

spending_summary_df["Average Math Score"] = spending_summary_df["Average Math Score"].map("{:,.2f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:,.2f}".format)
spending_summary_df["% Passing Math"] = spending_summary_df["% Passing Math"].map("{:,.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:,.2f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:,.2f}".format)
spending_summary_df
```
